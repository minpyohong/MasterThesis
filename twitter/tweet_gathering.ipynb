```python
import GetOldTweets3 as got
import datetime
import time
import pandas as pd
```

```python
days_range = []

query = 'bitcoin'

#query = 'ethereum'

#start = datetime.datetime.strptime("2020-5-8", "%Y-%m-%d")
#end = datetime.datetime.strptime("2020-5-12", "%Y-%m-%d")

start = datetime.datetime.strptime("2020-4-21", "%Y-%m-%d")
end = datetime.datetime.strptime("2020-5-7", "%Y-%m-%d")

#start = datetime.datetime.strptime("2017-3-19", "%Y-%m-%d")
#end = datetime.datetime.strptime("2017-6-20", "%Y-%m-%d")
#start = datetime.datetime.strptime("2017-9-15", "%Y-%m-%d")
#end = datetime.datetime.strptime("2017-12-26", "%Y-%m-%d")
#start = datetime.datetime.strptime("2018-1-7", "%Y-%m-%d")
#end = datetime.datetime.strptime("2018-2-5", "%Y-%m-%d")
#start = datetime.datetime.strptime("2018-9-9", "%Y-%m-%d")
#end = datetime.datetime.strptime("2018-11-18", "%Y-%m-%d")
#start = datetime.datetime.strptime("2018-11-19", "%Y-%m-%d")
#end = datetime.datetime.strptime("2018-12-14", "%Y-%m-%d")
#start = datetime.datetime.strptime("2019-4-2", "%Y-%m-%d")
#end = datetime.datetime.strptime("2019-6-26", "%Y-%m-%d")
#start = datetime.datetime.strptime("2019-7-10", "%Y-%m-%d")
#end = datetime.datetime.strptime("2019-7-16", "%Y-%m-%d")
#start = datetime.datetime.strptime("2019-7-31", "%Y-%m-%d")
#end = datetime.datetime.strptime("2019-8-8", "%Y-%m-%d")
#start = datetime.datetime.strptime("2019-8-10", "%Y-%m-%d")
#end = datetime.datetime.strptime("2019-8-14", "%Y-%m-%d")
#start = datetime.datetime.strptime("2019-9-23", "%Y-%m-%d")
#end = datetime.datetime.strptime("2019-9-26", "%Y-%m-%d")
#start = datetime.datetime.strptime("2019-10-25", "%Y-%m-%d")
#end = datetime.datetime.strptime("2019-10-27", "%Y-%m-%d")
#start = datetime.datetime.strptime("2019-11-8", "%Y-%m-%d")
#end = datetime.datetime.strptime("2019-11-24", "%Y-%m-%d")
#start = datetime.datetime.strptime("2020-1-3", "%Y-%m-%d")
#end = datetime.datetime.strptime("2020-2-11", "%Y-%m-%d")
#start = datetime.datetime.strptime("2020-3-8", "%Y-%m-%d")
#end = datetime.datetime.strptime("2020-3-30", "%Y-%m-%d")
date_generated = [start + datetime.timedelta(days=x) for x in range(0, (end-start).days)]

for date in date_generated:
    days_range.append(date.strftime("%Y-%m-%d"))

print("=== 설정된 트윗 수집 기간은 {} 에서 {} 까지 입니다 ===".format(days_range[0], days_range[-1]))
print("=== 총 {}일 간의 데이터 수집 ===".format(len(days_range)))

# 수집 기간 맞추기
start_date = days_range[0]
end_date = (datetime.datetime.strptime(days_range[-1], "%Y-%m-%d")
            + datetime.timedelta(days=1)).strftime("%Y-%m-%d")
```

```python
header = ['period','username','date','retweets','favorites','hashtags','text']
tweetSet = pd.DataFrame(columns=header)
```

```python
data = pd.read_csv('influencer.csv', error_bad_lines=False)
```

```python
data['TweetAccount'] = data['TweetAccount'].apply(lambda x : x.replace('@',''))
```

```python
data
```

```python
i=0
for j in range(0,data['No'].count()):
    #print(data.loc[i,'TweetAccount'])
    tweet_account = data.loc[j,'TweetAccount']
    tweetCriteria = got.manager.TweetCriteria().setQuerySearch(query)\
                                           .setSince(start_date)\
                                           .setUntil(end_date)\
                                           .setUsername(tweet_account)\
                                           .setMaxTweets(-1)
    print("Collecting data start.. from {} to {}".format(days_range[0], days_range[-1]))
    start_time = time.time()

    tweet = got.manager.TweetManager.getTweets(tweetCriteria)
    #print("Collecting data end.. {0:0.2f} Minutes".format((time.time() - start_time)/60))
    print(tweet_account,"=== Total num of tweets is {} ===".format(len(tweet))) 
    for tweet_contents in tweet :
        i = i+1
        
        #tweetSet.loc[i,'period'] = 202022
        tweetSet.loc[i,'period'] = 202012
        #tweetSet.loc[i,'period'] = 201711
        #tweetSet.loc[i,'period'] = 201712
        #tweetSet.loc[i,'period'] = 201821
        #tweetSet.loc[i,'period'] = 201831
        #tweetSet.loc[i,'period'] = 201822
        #tweetSet.loc[i,'period'] = 201911
        #tweetSet.loc[i,'period'] = 201921
        #tweetSet.loc[i,'period'] = 201912
        #tweetSet.loc[i,'period'] = 201922
        #tweetSet.loc[i,'period'] = 201923
        #tweetSet.loc[i,'period'] = 201913
        #tweetSet.loc[i,'period'] = 201924
        #tweetSet.loc[i,'period'] = 202011
        #tweetSet.loc[i,'period'] = 202021
        tweetSet.loc[i,'username'] = tweet_contents.username
        tweetSet.loc[i,'date'] = tweet_contents.date
        tweetSet.loc[i,'retweets'] = tweet_contents.retweets
        tweetSet.loc[i,'favorites'] = tweet_contents.favorites
        tweetSet.loc[i,'hashtags'] = tweet_contents.hashtags
        tweetSet.loc[i,'text'] = tweet_contents.text
        #print(str(i)+' '+tweet_contents.username+' '+str(tweet_contents.date) )
        
```
